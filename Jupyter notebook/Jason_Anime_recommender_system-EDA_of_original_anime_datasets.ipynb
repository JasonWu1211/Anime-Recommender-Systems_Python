## Import libraries

```python
#standard tools
import pandas as pd
from scipy.stats import ttest_rel,probplot
import numpy as np
from matplotlib import pyplot as plt
import re
import glob
import seaborn as sns
from sklearn.model_selection import train_test_split
from wordcloud import WordCloud
import math
import copy as cp
from collections import defaultdict
```

## Load Anime & Anime_rating csv into pandas Dataframe

```python
#list of anime files
Alst=glob.glob('anime*') 

#load anime files into list of pandas df
Adf=[]
for i in Alst:
    Adf.append(pd.read_csv(i))

Anime_rating = Adf[1]
Anime = Adf[0]
```

### Overview of Anime rating data ( User Rating on Anime )

```python

## print first 10 row of raw rating data
print('## first 10 row of raw rating data ## \n\n',Anime_rating.head())
print('='*50,'\n\n')

## summary of the obeservations that is : Anime watched by user but no rating ('rating' == -1) 
print("## Summary of the obeservations that is : Anime watched by user but no rating ('rating' == -1) ## \n")
norating = Anime_rating[Anime_rating['rating'] == -1]
print('Number of observations: {} '.format(len(norating)))

for i in norating.nunique().index:
    print('Number of  Unique {0}s: {1}'.format(i.split('_')[0],norating.nunique()[i]))

print('='*50,'\n\n')

Anime_rating=Anime_rating[Anime_rating['rating']!= -1] #drop rows with no rating (rating = -1)



## print summary of rating data
print('## Summary of rating data after dropping the no rating rows ## \n')
print('Number of observations: {} '.format(len(Anime_rating)))


for i in Anime_rating.nunique().index:
    print('Number of Unique {0}s: {1}'.format(i.split('_')[0],Anime_rating.nunique()[i]))

print('*rating scales: {}'.format(np.sort(Anime_rating['rating'].unique())))

print('Density: {}'.format(len(Anime_rating)/(Anime_rating.nunique()[0]*Anime_rating.nunique()[1])))

print('='*50)    
print('\n\n## first 10 rows of the data without no rating ## \n')
print(Anime_rating.head())
print('='*50)


## check missing vlaue
print('\n\n## Number of Missing values on rating data ##\n')
print(Anime_rating.isna().sum())
print('='*50)
```

### EDA with Graph

#### __Anime rating data__

##### *__Rating Levels__*

```python
## Frequecy of Rating ##

#df - summay of rating data rounded 2 decimal

rat_df=pd.DataFrame(Anime_rating.rating.describe().apply(lambda x : round(x,2)))


#plots
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.distplot(Anime_rating.iloc[:,2],hist=True,kde=False,hist_kws={'align': 'left'},axlabel='rating',bins=list(range(1,12)))
plt.ylabel('Frequency')
plt.title('Rating Distribution',fontsize=15)
plt.xticks(range(11))

plt.subplot(2,2,2)
plt.table(cellText = rat_df.values,rowLabels=rat_df.index,loc='center')
plt.title('Summary Statistic of Rating',fontsize=15)
plt.axis('off')


plt.tight_layout()

plt.show()
```

##### *__User Rating Distribution__*

```python
## Frequecy of the number of time that user rated ##

# df- summay of Number of time that user rated

u_rat_fre=Anime_rating.groupby('user_id')['rating'].count()

u_rat_df=pd.DataFrame(u_rat_fre.describe())

#plots
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)

        
sns.distplot(u_rat_fre,hist=False,kde_kws={'cumulative': True},axlabel='Number of time that user rated')
plt.title('CDF of number of user rating',fontsize=12)
plt.ylabel('Cumulative Probability')

plt.subplot(2,2,2)
plt.table(cellText = u_rat_df.values,rowLabels=u_rat_df.index,loc='center')
plt.title('Summary Statistic of Number of time that anime user rated',fontsize=12)
plt.axis('off')


plt.subplot(2,2,3)

        
sns.distplot(u_rat_fre,hist=True,kde=False,hist_kws={'log': True},axlabel='Number of time that user rated')
plt.title('User Rating Distribution',fontsize=12)
plt.ylabel('Frequency in log10')

plt.tight_layout()

plt.show()
```

##### *__Item Rating Distribution__*

```python
## Frequecy of the number of time that anime being rated ##

#summay of Number of time that anime being rated

a_rat_fre=Anime_rating.groupby('anime_id')['rating'].count()

a_rat_df=pd.DataFrame(a_rat_fre.describe())

#plots
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)

        
sns.distplot(a_rat_fre,hist=False,kde_kws={'cumulative': True},axlabel='Number of time that anime being rated')
plt.title('CDF of number of anime rating',fontsize=12)
plt.ylabel('Cumulative Probability')

plt.subplot(2,2,2)
plt.table(cellText = a_rat_df.values,rowLabels=a_rat_df.index,loc='center')
plt.title('Summary Statistic of Number of time that anime being rated',fontsize=12)
plt.axis('off')


plt.subplot(2,2,3)

        
sns.distplot(a_rat_fre,hist=True,kde=False,hist_kws={'log': True},axlabel='Number of time that anime being rated')
plt.title('Item Rating Distribution',fontsize=12)
plt.ylabel('Frequency in log10')


plt.tight_layout()

plt.show()
```

##### *__Most rated anime__*

```python
## visulizaing most rated anime : word cloud ##

#anime frequency df
Anime_fre=Anime_rating.groupby(Anime_rating.columns[1])['rating'].count().sort_values(ascending = False)
Anime_fre_df=pd.DataFrame(Anime_fre)
Anime_fre_df.columns=['count']
#anime name df
Aid_name=Anime.iloc[:,0:2].set_index('anime_id')

#inner join to get df with name, frequecny
Aname_fre=Aid_name.join(Anime_fre_df,how='inner')

#Top ten most popular anime
Top_ten = Aname_fre.sort_values('count',ascending=False).head(10)

#Aname_fre_dic
Aname_fre_dic={r['name']:r['count'] for _,r in Aname_fre.iterrows()}


#generate word cloud
wordCloud = WordCloud(mode='RGBA',background_color=None)



#plots
plt.figure(figsize=(15,15))

plt.subplot(2,2,1)
name_im= wordCloud.generate_from_frequencies(Aname_fre_dic)
plt.imshow(name_im,interpolation="bilinear")
plt.title('WordCloud: Most Frequenctly Rated Anime',fontsize=15)

plt.axis('off')
      
plt.subplot(2,2,2)
plt.table(cellText = Top_ten.values,rowLabels=Top_ten.index,colLabels=Top_ten.columns,loc='center')
plt.title('Top 10 most popular animes',fontsize=15)
plt.axis('off')
              
plt.tight_layout()

plt.show()
```

##### *__People who give perfect rating for everything vs People who give the lowest rating for everything__*

```python
#average rating by user
Ave_r_by_user=Anime_rating.groupby('user_id')['rating'].mean()

#people who only give perfect rating
u_10=Ave_r_by_user[Ave_r_by_user==10]
u_10_df=pd.DataFrame(u_10).reset_index()

#people who only give lowest rating
u_1=Ave_r_by_user[Ave_r_by_user==1]
u_1_df=pd.DataFrame(u_1).reset_index()
```

```python
#inner join to get user id, count ,and rating for 
#people who only give perfect rating & who only give lowest rating

#user rating count df
u_rat_fre_df=pd.DataFrame(u_rat_fre).reset_index()
u_rat_fre_df.columns=['user_id','count']


perfect_user_count_sorted = pd.merge(u_rat_fre_df,u_10_df,on='user_id').sort_values('count',ascending=False)
lowest_user_count_sorted = pd.merge(u_rat_fre_df,u_1_df,on='user_id').sort_values('count',ascending=False)
```

```python
print('Number of User who only give perfect rating in user dataset: {}'.format(u_10.count()))
print('Number of User who only give the lowest rating in user dataset: {}'.format(u_1.count()))
```

```python
#Plots : distribution of user rating count -perfect & the worst
plt.figure(figsize=(15,15))
plt.subplot(2,2,1)
sns.distplot(perfect_user_count_sorted['count'],hist_kws={'normed' : False,'log':True},kde=False)
plt.title('Perfect User Rating Distribution',fontsize=15)
plt.ylabel('Frequency in log10')
plt.xlabel('Number of rating')
plt.subplot(2,2,2)
sns.distplot(lowest_user_count_sorted['count'],hist_kws={'normed' : False,'align': 'left'},kde=False,bins=(1,2,3,4))
plt.title('worst User Rating Distribution',fontsize=15)
plt.ylabel('Frequency')
plt.xlabel('Number of rating')
plt.xticks((1,2,3))
plt.show()
```

__*Comment__:

lots of perfect rater and even if most of them only have rated less than 100 items. There are still 59  of them have rated over 100 animes

not many worst rater and all of them only have rated few animes
